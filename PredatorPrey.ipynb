# PredatorPrey

### Program to run simulation of the predator-prey relationship

```python
# PredatorPrey.ipynb

from time import time
from PredatorPrey import PredatorPrey_RK2, PredatorPrey_RK4 
import matplotlib.pyplot as plt            #Height (in inches) for the figures.
```

```python
# Measure time for RK-2
start_time = time()
timeRK2, predatorRK2, preyRK2 = PredatorPrey_RK2()
end_time = time()
rk2_execution_time = end_time - start_time
```

```python
# Measure time for RK-4
start_time = time()
timeRK4, predatorRK4, preyRK4 = PredatorPrey_RK4()
end_time = time()
rk4_execution_time = end_time - start_time
```

```python
# Plotting RK-2 and RK-4 on the same graph
plt.figure(figsize=(10, 6))
plt.plot(timeRK2, predatorRK2, 'b-', label='Predator RK-2')
plt.plot(timeRK2, preyRK2, 'r-', label='Prey RK-2')
plt.plot(timeRK4, predatorRK4, 'g--', label='Predator RK-4')
plt.plot(timeRK4, preyRK4, 'y--', label='Prey RK-4')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Predator and Prey Populations vs Time (RK-2 and RK-4)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Output execution times
print("RK-2 Execution Time:", rk2_execution_time)
print("RK-4 Execution Time:", rk4_execution_time)
```
