# sharkCompetition

### Species competition between white tip sharks (WTS) and black tip sharks (BTS)

```python
#This code performs the necessary steps to prepare to plot.
from pylab import *     #Import plotting module matplotlib as well as other modules
                        #into the global namespace.

#Set the default plot to be inline rather than a new window.
%matplotlib inline

import inspect  #This allows us to print the source code in this notebook.
from sharkCompetition import sharkCompetition
import time
```

```python

goldenratio=1/2*(1+sqrt(5))        #The next few lines are used for the size of plots
fsx=7                              #Width (in inches) for the figures.
fsy=fsx/goldenratio                #Height (in inches) for the figures.
```

```python
# Simulate with RK-2
start_time = time.time()
timeLst_rk2, WTSLst_rk2, BTSLst_rk2 = sharkCompetition(method='RK2')
rk2_execution_time = time.time() - start_time
```

```python
# Simulate with RK-4
start_time = time.time()
timeLst_rk4, WTSLst_rk4, BTSLst_rk4 = sharkCompetition(method='RK4')
rk4_execution_time = time.time() - start_time
```

```python
# Graph
figure(figsize=(fsx, fsy))
plot(timeLst_rk2, WTSLst_rk2, 'b-', label='WTS Population (RK2)')
plot(timeLst_rk2, BTSLst_rk2, 'r-', label='BTS Population (RK2)')
plot(timeLst_rk4, WTSLst_rk4, 'g--', label='WTS Population (RK4)')
plot(timeLst_rk4, BTSLst_rk4, 'm--', label='BTS Population (RK4)')
legend(loc='upper center')
xlabel('Time')
ylabel('Population')
title('WTS and BTS Populations vs Time (RK2 and RK4)')

print("RK-2 execution time:", rk2_execution_time)
print("RK-4 execution time:", rk4_execution_time)

show()
```
